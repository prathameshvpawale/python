```python
# Program to demonstrate standard Data Types

def main():
    # 1. Integer (whole numbers)
    age = 25
    print(f"Value: {age}, Type: {type(age)}")

    # 2. Float (decimal numbers)
    price = 19.99
    print(f"Value: {price}, Type: {type(price)}")

    # 3. String (text)
    name = "Alice"
    print(f"Value: {name}, Type: {type(name)}")

    # 4. Boolean (True/False)
    is_student = True
    print(f"Value: {is_student}, Type: {type(is_student)}")

    # 5. List (Ordered, mutable collection)
    scores = [90, 85, 88]
    print(f"Value: {scores}, Type: {type(scores)}")

    # 6. Dictionary (Key-Value pairs)
    person = {"name": "Bob", "role": "Admin"}
    print(f"Value: {person}, Type: {type(person)}")

if __name__ == "__main__":
    main()
```

```python
# Program to demonstrate Decision Making (Control Flow)

def check_number(num):
    # The code indented under 'if', 'elif', and 'else' forms a BLOCK.
    
    if num > 0:
        print(f"{num} is a Positive number.")
    elif num < 0:
        print(f"{num} is a Negative number.")
    else:
        print("The number is Zero.")

# Testing the function
check_number(10)
check_number(-5)
check_number(0)
```

```python
# Program to demonstrate 'For' Loops

print("--- Fruit List ---")
fruits = ["Apple", "Banana", "Cherry"]

# 'fruit' is a temporary variable representing the current item
for fruit in fruits:
    print(f"I like {fruit}")  # This block repeats for every item in the list
```

```python
# Program to demonstrate 'While' Loops

print("\n--- Countdown ---")
count = 5

while count > 0:
    print(f"Count is: {count}")
    count = count - 1  # Decrement the counter

print("Blast off!")
```

```python
# Program to demonstrate Break and Continue

print("--- Loop Control ---")

for i in range(1, 10):
    if i == 3:
        print("Skipping 3 (Continue)")
        continue  # Skips the print statement below for i=3
    
    if i == 7:
        print("Stopping at 7 (Break)")
        break  # Stops the loop entirely
        
    print(f"Processing number: {i}")
```
