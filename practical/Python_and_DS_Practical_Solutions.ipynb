## 3. Singly Linked List

```
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=' -> ')
            temp = temp.next
        print('None')
```

## 4. Recursive Binary Search

```
def binary_search(arr, low, high, key):
    if low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif key < arr[mid]:
            return binary_search(arr, low, mid - 1, key)
        else:
            return binary_search(arr, mid + 1, high, key)
    return -1

arr = [1, 3, 5, 7, 9]
print(binary_search(arr, 0, len(arr)-1, 7))
```

## 5. Sorting – Bubble Sort

```
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

arr = [64, 34, 25, 12, 22]
bubble_sort(arr)
print(arr)
```

## 6. Graph – Adjacency List & DFS

```
graph = {}

def add_edge(u, v):
    graph.setdefault(u, []).append(v)

def dfs(node, visited):
    visited.add(node)
    print(node, end=' ')
    for n in graph.get(node, []):
        if n not in visited:
            dfs(n, visited)

add_edge(0, 1)
add_edge(0, 2)
add_edge(1, 3)

dfs(0, set())
```
